```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import gamma
```

```python
samples = 7 # this is the parameter n for chi_n_squared distribution

trials = 9000 # this is the number of times we sample the chi_n_squared distribution
plot_range = 4 * samples + 1 # plot range for chi_n_squared distribution
```

```python
data = []
frequency = np.zeros(plot_range)
for i in range(trials):
    X = np.random.randn(samples)
    Xsquared = X**2
    datum = sum(Xsquared)
    data.append(datum)
    if (int(datum) < plot_range):
        frequency[int(datum)] += 1
frequency /= trials
plt.plot(frequency, 'r.')
x = np.arange(0, plot_range, 0.1)
y = (x**(samples/2 - 1) * np.exp(-x/2)) / (2**(samples/2)* gamma(samples/2))
plt.plot(x, y)
```
